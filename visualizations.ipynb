```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
crit_loss_dcgan_batch_100 = np.load("dcgan_batch_100/critic_loss-1000.npy")
crit_loss_dcgan_batch_100.shape
```

```python

iterations = np.arange(1, len(crit_loss_dcgan_batch_100) + 1)

# Plotting every 100th value
plt.figure(figsize=(10, 6))
sns.lineplot(x=iterations[:200], y=crit_loss_dcgan_batch_100[:200])
plt.title('DCGAN discriminator Loss Over Checkpoints')
plt.xlabel('Iterations')
plt.ylabel('Loss Value')
plt.xticks(np.arange(0, 201, 10))  # Show every 100th value on x-axis up to 1000 iterations
plt.grid(True)
plt.suptitle("Plot by A.Sahakyan")
plt.show()
```
