```python
import numpy as np
import matplotlib.pyplot as plt

def generate_centered_gaussian_kernel(size=3, sigma=1.0):
    """Generate a 2D Gaussian kernel with a centered hotspot."""
    ax = np.linspace(-(size - 1) / 2., (size - 1) / 2., size)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-0.5 * (np.square(xx) + np.square(yy)) / np.square(sigma))
    return kernel / np.sum(kernel)

# Generate a centered 3x3 Gaussian kernel
size = 3
sigma = 1.0
centered_gaussian_kernel = generate_centered_gaussian_kernel(size, sigma)

# Plot the 3x3 Gaussian kernel as a heatmap
plt.figure(figsize=(6, 6))
plt.imshow(centered_gaussian_kernel, cmap='viridis', interpolation='none')
plt.colorbar()
plt.title('Centered 3x3 Gaussian Kernel')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(False)
plt.show()
```
