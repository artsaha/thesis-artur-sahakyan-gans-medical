```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import truncnorm, norm

# Set seed for reproducibility
np.random.seed(0)

# Define parameters for the truncated normal distribution
a, b = -2, 2  # Lower and upper bounds
mu, sigma = 0, 1  # Mean and standard deviation

# Generate random samples from the truncated normal distribution
samples_truncated = truncnorm.rvs((a - mu) / sigma, (b - mu) / sigma, loc=mu, scale=sigma, size=10000)

# Generate random samples from the normal distribution
samples_normal = np.random.normal(mu, sigma, 10000)

# Define the range for plotting the curves
x = np.linspace(-3, 3, 1000)

# Calculate the probability density function (PDF) for the normal distribution
pdf_normal = norm.pdf(x, mu, sigma)

# Calculate the PDF for the truncated normal distribution
pdf_truncated = truncnorm.pdf(x, (a - mu) / sigma, (b - mu) / sigma, loc=mu, scale=sigma)

# Adjust the PDF values of the truncated distribution outside the bounds
pdf_truncated[x < a] = pdf_truncated[a]
pdf_truncated[x > b] = pdf_truncated[b]

# Plot the normal distribution
plt.plot(x, pdf_normal, label='Normal Distribution', color='skyblue')

# Plot the truncated normal distribution
plt.plot(x, pdf_truncated, label='Truncated Normal Distribution', color='salmon')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Truncated Normal Distribution')
plt.legend()

plt.savefig('visual/architecture/truncated_normal.png', dpi=300)
# Show plot
plt.show()
```
